```python

def is_safe(graph, color, vertex, c):
    """Check if it's safe to assign color c to vertex."""
    for neighbor in graph[vertex]:
        if color[neighbor] == c:
            return False
    return True

def graph_coloring_util(graph, colors, color, vertex):
    """Utilize backtracking to solve the graph coloring problem."""
    if vertex == len(graph):
        return True

    for c in colors:
        if is_safe(graph, color, vertex, c):
            color[vertex] = c
            if graph_coloring_util(graph, colors, color, vertex + 1):
                return True
            color[vertex] = None

    return False

def graph_coloring(graph, colors):
    """Main function to solve the problem."""
    color = [None] * len(graph)

    if graph_coloring_util(graph, colors, color, 0):
        return color
    else:
        return None

if __name__ == "__main__":
    graph = {
        0: [1, 2],
        1: [0, 2, 3],
        2: [0, 1],
        3: [1]
    }

    colors = ['R', 'G', 'Y', 'P']
    result = graph_coloring(graph, colors)

    if result:
        print("Coloring of the graph:", result)
    else:
        print("No solution exists.")
```
